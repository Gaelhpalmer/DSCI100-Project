## Data Preprocessing ##

#### Splitting ####

```r
# Seed set to allow reproduction of analysis. 
# Splitting performed to form training and evaluation sets to be used in analysis.
# Further explanation on splitting mechanisms and proportion selection are provided in methodology and results sections. 

set.seed(seed)
stroke_split <- initial_split(stroke_clean, prop = 0.75, strata = stroke)
stroke_train <- training(stroke_split)
stroke_test <- testing(stroke_split)
```

#### Balancing ####

Because of the highly imbalanced nature of the data we are working with (as strokes are a relatively rare occurrance) balancing was necessary. Downsampling was selected as the most appropriate method.

```r
# Seed set to allow reproduction of analysis. 
# Further explanation on balancing is provided in the methodology and results sections.

set.seed(seed)
ds_recipe <- recipe(stroke ~ ., data = stroke_train) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors()) %>%
    step_downsample(stroke, under_ratio = 1, skip = FALSE) %>%
    prep()
dsampled_stroke_train <- bake(ds_recipe, stroke_train)

# Downsampling was only applied to the training set, as it was what will be used in the model training. 

# After downsampling, it can be seen that the training set has equal proportions of stroke/no stroke.
print("Table 8: Occurrence of Stroke After Balancing")
dsampled_stroke_train %>%
    group_by(stroke) %>%
    summarize(n = n())
```

#### Forward Selection of Predictor Variables ####
Forward prediction was performed to see which variables were optimal predictors for stroke. While this was not necessary, we opted to include it in our analysis to ensure we achieved the best model possible.

```r
# Firstly, a subset was created using the scaled and downsampled training data.

stroke_subset <- dsampled_stroke_train %>%
    filter(!is.na(bmi))

# A set of all available predictors was created from the subset.

names <- colnames(stroke_subset %>% select(-stroke))
names
```

```r
# An empty tibble was created to store results.

accuracies <- tibble(size = integer(), 
                     model_string = character(), 
                     accuracy = numeric())

# The total number of predictors was stored.

n_total <- length(names)

# An empty vector was created to store the selected predictors.

selected <- c()
```

```r
# A model specification was created using tune() for neighbors.

knn_spec <- nearest_neighbor(weight_func = "rectangular", 
                             neighbors = tune()) %>%
     set_engine("kknn") %>%
     set_mode("classification")

# A 5-fold cross validation object was created.

stroke_vfold <- vfold_cv(stroke_subset, v = 5, strata = stroke)

# A grid values tibble was created, using minmal values to prevent long running times.
gridvals <- tibble(neighbors = seq(from = 1, to = 20, by = 3))
```

```r
set.seed(seed)
# Note: takes about 1m to run
# for every size from 1 to the total number of predictors
for (i in 1:n_total) {
    print("start")
    # for the predictors not yet added
    accs <- list()
    models <- list()
    for (j in 1:length(names)) {
        # a model string is created for this combination of predictors
        preds_new <- c(selected, names[[j]])
        model_string <- paste("stroke", "~", paste(preds_new, collapse="+"))
        print(model_string)
        
        # a recipe is created from the model string
        stroke_recipe <- recipe(as.formula(model_string), 
                                data = stroke_subset) %>%
                          step_scale(all_predictors()) %>%
                          step_center(all_predictors())

        # the KNN classifier is tuned with these predictors, 
        # and accuracy for the best K is collected
        set.seed(seed)
        acc <- workflow() %>%
          add_recipe(stroke_recipe) %>%
          add_model(knn_spec) %>%
          tune_grid(resamples = stroke_vfold, grid = gridvals) %>%
          collect_metrics() %>%
          filter(.metric == "accuracy") %>%
          summarize(mx = max(mean))
        acc <- acc$mx %>% unlist()

        # add this result to the dataframe
        accs[[j]] <- acc
        models[[j]] <- model_string
    }
    set.seed(seed)
    print("unlist")
    print(unlist(accs))
    jstar <- which.max(unlist(accs))
    accuracies <- accuracies %>% 
      add_row(size = i, 
              model_string = models[[jstar]], 
              accuracy = accs[[jstar]])
    selected <- c(selected, names[[jstar]])
    names <- names[-jstar]
}
accuracies

print("Table 9: Results of Forward Selection")
```

```r
# A visualization of the accuracy as the number of predictors increases.

# While 3 is the most optimal number of predictors, we opted for 2 to make visualization more simple,
# and the difference in accuracy between 2 and 3 predictors was minimal.
options(repr.plot.width = 7, repr.plot.height = 7)
ggplot(accuracies, aes(x=size, y=accuracy)) +
    geom_line() +
    geom_point() +
    labs(caption = "Figure 7") +
    ggtitle("Number of Predictors vs ")


# Thus, age and average glucose levels will give the best model accuracy.
```

#### K Value Selection ####

```r
set.seed(seed)

# A recipe was created using downsampled training data with the optimal predictors selected.

stroke_recipe <- recipe(stroke ~ avg_glucose_level + age, data = dsampled_stroke_train) %>%
  step_scale(all_predictors()) %>%
  step_center(all_predictors())

# A cross-validation object was created using the downsampled training data.

stroke_vfold <- vfold_cv(dsampled_stroke_train, v = 5, strata = stroke)

# A grid value tibble of K values was created to be used in tuning.

# We chose the cap it at 75 and step it up by 3 to prevent the computational time from being too long, 
# while still giving a fairly accurate estimate of the optimal k value.

k_vals <- tibble(neighbors = seq(from = 1, to = 75, by = 3))

# A KNN classification model specification was created using tune().

knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) %>%
  set_engine("kknn") %>%
  set_mode("classification")

# A workflow was created to collect metrics on accuracy as K values were varied using the model specifications above.

knn_results <- workflow() %>%
  add_recipe(stroke_recipe) %>%
  add_model(knn_spec) %>%
  tune_grid(resamples = stroke_vfold, grid = k_vals) %>%
  collect_metrics() 
```

```r
# The optimal K value after cross-validation k-selection is displayed below, as well as a graph of K vs Accuracy.

accuracies <- knn_results %>% # Plot accuracies vs neighbours
  filter(.metric == "accuracy")
  ggplot(accuracies, aes(x=neighbors, y=mean)) +
  geom_line() +
  ggtitle("K-values vs Accuracy for Stroke Prediction Model") +
  xlab("Number of Neighbors (K)") +
  ylab("Accuracy") +
  labs(caption = "Figure 8") +
  theme(text = element_text(size=15))


optimal_neighbours <- accuracies %>%
  filter(mean == max(mean))
print("Table 10: Optimal Number of Neighbors")
optimal_neighbours
```

## Final Model ##
After predictor and k-value optimization, we are constructing a final model.

```r
set.seed(seed)

# A model specification was created using the optimal K of 10.

final_stroke_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 10) %>%
  set_engine("kknn") %>%
  set_mode("classification")

# A workflow was created, using the new specification. 

final_stroke_fit <- workflow() %>%
  add_recipe(stroke_recipe) %>%
  add_model(final_stroke_spec) %>%
  fit(data = dsampled_stroke_train)

final_stroke_fit
```

#### Accuracy Evaluation ####

```r
# First, testing data was standardized.
set.seed(seed)

standardize_recipe <- recipe(stroke ~ ., data = stroke_test) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors()) %>%
    prep()

stroke_test_standardized <- bake(standardize_recipe, stroke_test)
```

```r
stroke_test_predictions <- predict(final_stroke_fit, stroke_test_standardized) %>%
  bind_cols(stroke_test_standardized)


stroke_pred_metrics <- stroke_test_predictions %>%
  metrics(truth = stroke, estimate = .pred_class)
print("Table 11: Final Model Accuracy Metrics")
stroke_pred_metrics
```

#### Confusion Matrix ####

```r
# A confusion matrix was created to assess the nature of error in our model.

stroke_mat <- stroke_test_predictions %>% 
      conf_mat(truth = stroke, estimate = .pred_class) 

stroke_mat
print("         Figure 10")
```

When applied to previously unseen data, the model achieved an accuracy rate of ~70% and was more prevalent to type I error. 

#### Visualization of Model Classification Boundaries ####

```r
# In order to visualise the model we have created, we made a scatter plot with the original data points, along with shaded
# regions indicating which values would be predicted as stroke vs no stroke. 

age_grid <- seq(min(stroke_test_standardized$age), 
                max(stroke_test_standardized$age), 
                length.out = 100)
glu_grid <- seq(min(stroke_test_standardized$avg_glucose_level), 
                max(stroke_test_standardized$avg_glucose_level), 
                length.out = 100)
asgrid <- as_tibble(expand.grid(age = age_grid, 
                                avg_glucose_level = glu_grid))

knnPredGrid <- predict(final_stroke_fit, asgrid)

prediction_table <- bind_cols(knnPredGrid, asgrid) %>%
  rename(Class = .pred_class)


wkflw_plot <-
  ggplot() +
  geom_point(data = stroke_test_standardized, 
             mapping = aes(x = age, 
                           y = avg_glucose_level, 
                           color = stroke), 
             alpha = 0.75) +
  geom_point(data = prediction_table, 
             mapping = aes(x = age, 
                           y = avg_glucose_level, 
                           color = Class), 
             alpha = 0.04, 
             size = 5) +
  labs(color = "Stroke Occurance", 
       x = "Age (standardized)", 
       y = "Average Blood Glucose Levels (standardized)", 
       caption = "Figure 11") +
  scale_color_manual(labels = c("No Stroke", "Stroke"), 
                     values = c("orange2", "steelblue2")) +
  theme(text = element_text(size = 12))+ 
    ggtitle("Predicting the Occurance of Stroke Using Age and Blood Glucose Levels")

wkflw_plot
```

## Methodology and Rationale ##
First, exploratory data analysis was conducted to gain a better understanding of the selected data set. This analysis elucidated 3 key findings. First, after cleaning the data (converting appropriate categories to factors), it was clear that the data was already in a tidy format and no wrangling would be necessary. Secondly, the data contained  4861 non-stroke data points and only 249 stroke data points (a 95% to 5% ratio). This severe class imbalance meant that overclassification of the majority class was a concern, making balancing the data necessary to ensure a functioning model.  Finally, our visualizations (figures 2-6) showing different correlational relationships indicated that age and average glucose level were likely predictor variable candidates for our model.

Prior to training the model, the data needed to be preprocessed. First, a seed value was set in order to ensure the reproduction of our analysis was possible, while still maintaining random sampling. The data was then split into training and testing data subsets. A 75-25 training-testing split was used. This ratio was selected as to have enough data for the model to effectively and accurately learn, while also still leaving enough data points to evaluate the accuracy of the model. Then, to balance the data the technique of downsampling was used, whereby a random sample of data is taken ensuring that the majority class has the same proportional representation as the minority class. In our case, this meant having 189 data points for each ‘stroke’ and ‘no stroke’. Downsampling was favoured over upsampling as the upsampling technique available to us consisted of cloning minority class observations, which after splitting the data meant our model would be trained on points that when cross-validated would appear again, making our accuracy seem higher than its true value. During downsampling, the data was also scaled and centered. It should be noted that we only downsampled the training data because that is what would be used to train the model. 

Next, we needed to determine which, and how many, predictive variables we would use for the model. To do so, we elected to perform forward selection. We decided to use this method, as creating subsets of all the different combinations of predictors would take too much time, and forward selection would allow us to use fewer subsets of predictors to find a good estimate of the optimal set of predictors to maximize the accuracy of the final model. The 5 potential variables were age (numerical), hypertension (binary), heart disease (binary), average glucose level (numerical), and body mass index (numerical). First, we created models only using one of each of the 5 variables and then chose the model that gave the highest cross-validation accuracy. We used the predictor for that model to create more models, where one other predictor was also used in the model for each of these models, and again chose the model with the highest cross-validation accuracy. We continued this process of adding another predictor onto the model that gave the highest accuracy until we ended up with 5 models that gave the highest cross-validation accuracies in their group. We then graphed the accuracies of these models. The resulting graph (figure 7) showed that using age, average glucose levels, and hypertension would result in the highest accuracy (0.773) while using only age and average glucose levels gave the second-highest accuracy (0.759). Because this was only a difference of 0.014, and we were intent on being able to create a simple visual representation of the model, we elected to use only 2 variables (age and glucose levels). Trying to avoid overfitting the data to the model was an additional reason for selecting 2 rather than 3 variables. Moreover, previous research also indicated that both age and glucose were correlated with the risk of stroke, thus, using these variables as predictors was scientifically sound (see discussion section for details on research). 

To create the predictive model, the tidymodel's k-nearest neighbour classification algorithm was used (KNN). First, the appropriate K value was to be determined. We chose to perform cross-validation with the range of 1-75 neighbours and stepped the K values up by 3. We chose the maximum of 75 and stepped the K values up by 3 to prevent the computational time from being too long while still giving a fairly accurate estimate of the optimal K value. After running a 5 fold cross-validation test, and scaling the data of average glucose level and age so as to compare them without units, we generated a graph that plotted the potential K values against their respective mean accuracies (figure 8). The K value with the highest accuracy was 10. We deemed this to be an appropriate choice because the value was high enough that it could take into account outliers and slightly unusual data without ‘overfitting’, and low enough so as not to create an averaging effect, where the prediction is derived from whatever observation is most common within the entire dataset. The value of 10 was also not too computationally expensive, which a higher value may have been.

After this step, we were able to create a workflow using the optimized K value. In doing so, the downsampled data was applied to the model to build the classifier. With the model completed, we used the metrics function to generate and assess its accuracy. The “.estimate” column displays the accuracy, which was estimated here to be 70.5%. 

Finally, to visualize our final model, we created a scatter plot (figure 11) that showed both actual data points (opaque), as well as shading of how unknown data points would be classified. This was done by inputting a spectrum of values for both age and blood glucose into our model, and the outputs were graphed as points with low opacity.

Overall, our results show that we can predict if someone will have a stroke, using age and blood glucose levels, with an accuracy of 70.5%. Looking at the confusion matrix, we can interpret that there are 850 true negatives (predicted no stroke, actually no stroke), 9 false negatives (predicted no stroke, actually stroke), 367 false positives (predicted stroke, actually no stroke), and 51 true positives (predicted stroke, actually stroke). 

## Discussion ##
Our final model used age and blood glucose levels in order to predict the incidence of stroke. A KNN algorithm was used to create the model, with 10 neighbours used to classify an unknown point. The accuracy of this model at predicting the incidence of stroke was 70.6%. Based on this accuracy, the model does have the potential for clinical applications in identifying whether closer attention should be paid to certain patients in preventing strokes. However, it should not be used as a decision-maker for whether or not major action such as medication or surgical procedures should be taken to prevent stroke, as the accuracy is not high enough. The repercussions of a false positive prediction in the intended application would be minimal, with a possibility of undue mental stress. In stark contrast, a false negative could result in mismanagement of health and no intervention and may result in the occurrence of an otherwise preventable stroke. Fortunately, based on our model’s confusion matrix (figure 10), only 0.7% of predictions were false negatives.

It was not a surprise that the two variables used as predictors were age and average glucose level, and this was what we had initially predicted would be used in our model. The correlation between both of these variables and stroke occurrence is well supported by consequently-referenced literature. For example, the risk of stroke doubles every 10 years after the age of 45 (Kelly-Hayes, 2010). This is because as we age, arteries become narrower and firmer, and we risk developing atherosclerosis, creating the conditions for stroke to occur (Stroke Association). With respect to the impact of blood glucose levels, it has been shown that long-term elevated blood glucose levels can damage blood vessels, leading to an increased risk of stroke. Furthermore, in a meta-analysis of studies involving over 2.5 million participants, it was established that there is a positive correlation between fasting blood glucose and risk of stroke, with the optimal fasting blood glucose level identified as 70-100 mg/dL (Shi et al, 2021). 

In a global context, our model and findings could have clinical applications in stroke prevention by proactively monitoring these correlated predictors in given individuals. This way, those who are susceptible can take the necessary medical precautions and treatments to avoid a stroke altogether, such as medications or changes in lifestyle. Seeing as strokes are the second leading cause of death worldwide, further studies on this topic have the potential to save a significant amount of lives.

It is also relevant to question the effectiveness of our model, and if there is any potential for improvement within this dataset. For example, one might ask if we would have achieved more accurate results by splitting the data by sex, ethnicity, or any binary observations which may be correlated to other parts of the dataset? In doing so, would we have a more effective model which accounts for these alternative predictors? In addition to this, we call into question if it is possible to use this dataset to create a model that is not only a binary classifier but evaluates the actual probability of whether or not a given individual will suffer from a stroke, and what specific precautions could be taken based on given health indicators. After all, the medical world rarely deals with absolutes, and each of the predictors used in this research are only indicators of someone’s susceptibility to experiencing a stroke rather than a guarantee of such an outcome.
